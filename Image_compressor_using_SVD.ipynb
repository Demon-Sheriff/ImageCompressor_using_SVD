`The image comes out to be a 3 dimensional vector 
 #the first dimension is the width
 #the second dimension is the height
 #the third dimension is the color channels
 `
 ``# In simpler terms we convert the image such that it will no longer be a 3d image it will become a 2d image``

```python
A = grayscaleforimg3 # store the image as a 2d matrix
U,Sigma,Vt = np.linalg.svd(A,full_matrices=False)
print(U.shape,Sigma.shape,Vt.shape)
print(Sigma)
```

```python
Sigma = np.diag(Sigma)
print(Sigma.shape)
Sigma
print(U.shape,Sigma.shape,Vt.shape)
```

```python
A = grayscaleforimg3 # store the image as a 2d matrix
U,Sigma,Vt = np.linalg.svd(A,full_matrices=False)
# print(U.shape,Sigma.shape,Vt.shape)
# print(Sigma)

Sigma = np.diag(Sigma)
# print(Sigma.shape)
# print(U.shape,Sigma.shape,Vt.shape)

# writing a function to compress image
import matplotlib.pyplot as plt

def compress_image(r):
    # subplot for the original image
    plt.subplot(1, 2, 1)
    store_img = plt.imshow(grayscaleforimg3)
    store_img.set_cmap('gray')
    plt.title('Original')

    # subplot for the compressed image
    img_compressed = U[:, :r] @ Sigma[:r, :r] @ Vt[:r, :]
    plt.subplot(1, 2, 2)
    img = plt.imshow(img_compressed)
    img.set_cmap('gray')
    plt.title(f'Compressed at r = {r}')

    plt.show()

compress_image(50)
```
